```python
# Image pre-processor cleans a given color image and produces a binary image output

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# taking input image
tbpg = cv2.imread("input.jpg")
if tbpg.all() == None:
    print("image not found")
    
# changing to standard rgb color
tbpg = cv2.cvtColor(tbpg, cv2.COLOR_BGR2RGB)
# gray-scaling
tbpg = cv2.cvtColor(tbpg, cv2.COLOR_RGB2GRAY)
# sharpening.
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
tbpg = cv2.filter2D(src=tbpg, ddepth=-1, kernel=kernel)
# denoising
tbpg = cv2.fastNlMeansDenoising(tbpg, h=40)
# binarization
tbpg = cv2.adaptiveThreshold(tbpg,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

# display result
plt.imshow(tbpg, cmap="binary", vmin=0, vmax=255)

# save result
cv2.imwrite('output.jpg', tbpg)
```
